```python
# Retrieve the title of the book
titles = []

for page in range(1, 11):  # 10 pages in total
    url = f"{base_url}?page={page}"
    #response = requests.get(url)
    #soup = BeautifulSoup(response.content, 'html.parser')

    for row in soup.select('table.tableList tr'):
        title = row.select_one('td:nth-child(3) > a > span')
        if title:
            titles.append(title.text.strip())

    time.sleep(1)  # Sleep to avoid rate limiting

# Do something with the books list, like printing or saving to a file
for book in titles:
    print(titles)
```

```python
# Retrieve the images of the book
image_urls = []

for page in range(1, 11):  #  there are 10 pages
    url = f"{base_url}?page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    for row in soup.select('table.tableList tr'):
        img_tag = row.select_one('td:nth-child(2) > div > a > img')
        if img_tag and img_tag.get('src'):
            img_url = img_tag['src']
            image_urls.append(img_url)

    time.sleep(1)  # Sleep to avoid rate limiting

# Print or process the collected image URLs
for url in image_urls:
    print(url)
```

```python
#put all the infomation to the dataframe
data_book = pd.DataFrame({"authors":books,
                       "original_title":titles,"image_url":image_urls
                      })
data_book
```
